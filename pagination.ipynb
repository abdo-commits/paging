```python
# We build a function that will keep paginating through the api endpoint but once we reach a certain data we stop
def paginate_api(api_endpoint, page_size=100, page_number=1, created="2020-01-01"):
    """
    Function to paginate through the api endpoint
    :param api_endpoint: The api endpoint to paginate through
    :param page_size: The number of results per page
    """
    
    # We initialize a list to store the results
    results = []
    while True:
        response = requests.get(api_endpoint.format(page_size=page_size, page_number=page_number))
        results.extend(response.json())
        response.raise_for_status()
        page_number += 1
        if len(response.json()) < page_size or response.json()["created"] < created:
            break
        
    return results
        
```

```python
paginate_api("https://rickandmortyapi.com/api/character?page=")
```

```python
import requests
from concurrent.futures import ThreadPoolExecutor, as_completed

# Function to fetch the data
def fetch(page):
    response = requests.get(f"https://rickandmortyapi.com/api/character?page={page}")
    return response.json()

# Function to handle the condition checking and fetching
def main():
    page = 1
    condition_met = False

    while not condition_met:
        # Start fetching multiple pages in parallel
        with ThreadPoolExecutor(max_workers=5) as executor:
            futures = {executor.submit(fetch, i) for i in range(page, page+5)}

            for future in as_completed(futures):
                result = future.result()

                # Check your condition here
                if 'something' in result:  # Replace 'something' with your condition
                    condition_met = True
                    break

        page += 5

main()
```
